```python
%matplotlib inline
from datetime import datetime, timedelta
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
latest_date_row = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest_date = latest_date_row[0]

# Calculate the date one year from the last date in data set.
latest_date = datetime.strptime(latest_date, '%Y-%m-%d')
one_year_ago = latest_date - timedelta(days=365)
# Perform a query to retrieve the data and precipitation scores
results = session.query(Measurement).filter(Measurement.date >= one_year_ago).all()
# Save the query results as a Pandas DataFrame. Explicitly set the column names

# Creating empty lists to store the values of each column
date = []
prcp = []

# Iterate over the results and extract the desired attributes
for measurement in results:
    date.append(measurement.date)
    prcp.append(measurement.prcp)
    
#Creating df
last_year = pd.DataFrame({'date': date, 'prcp': prcp})

# Print the DataFrame
print(last_year)

# Sort the dataframe by date
last_year_sorted = last_year.sort_values('date')

# Use Pandas Plotting with Matplotlib to plot the data
# Converting the 'date' column to datetime object
last_year_sorted['date'] = pd.to_datetime(last_year_sorted['date'])

# Setting up figure and axes
fig, ax = plt.subplots()

# Plotting line graph
ax.plot(last_year_sorted['date'], last_year_sorted['prcp'], linestyle='-', marker='o')

#labeling
ax.set_xlabel('Date')
fig.autofmt_xdate(rotation=45)

ax.set_ylabel('Precipitation')

ax.set_title('Precipitation Over Time')

# Display the graph
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
count = last_year_sorted['prcp'].count()
mean = last_year_sorted['prcp'].mean()
std = last_year_sorted['prcp'].std()
min_value = last_year_sorted['prcp'].min()
q1 = last_year_sorted['prcp'].quantile(0.25)
median = last_year_sorted['prcp'].median()
q3 = last_year_sorted['prcp'].quantile(0.75)
max_value = last_year_sorted['prcp'].max()
print("Count:", count)
print("Mean:", mean)
print("Standard Deviation:", std)
print("Minimum:", min_value)
print("25th Percentile (Q1):", q1)
print("Median (Q2):", median)
print("75th Percentile (Q3):", q3)
print("Maximum:", max_value)
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
# number of rows (stations)
station_count_query = session.query(func.count(Station.id))
station_count = station_count_query.scalar()

# Print the number of stations
print("Number of stations:", station_count)
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
station_query = session.query(Measurement.station, func.count(Measurement.station).label('row_count')).group_by(Measurement.station).order_by(func.count(Measurement.station).desc())

# Execute the query and fetch all the results
stations = station_query.all()

# Print the stations and their row counts in descending order
for station in stations:
    print("Station:", station.station, "Row Count:", station.row_count)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
station_id = 'USC00519281'

# Design a query to calculate the lowest, highest, and average temperature for the specified station
temperature_query = session.query(func.min(Measurement.tobs).label('lowest_temp'),
                                  func.max(Measurement.tobs).label('highest_temp'),
                                  func.avg(Measurement.tobs).label('avg_temp')).filter(Measurement.station == station_id)

# Execute the query and fetch the result
temp_result = temperature_query.one()

# Extract the temperature values
lowest_temp = temp_result.lowest_temp
highest_temp = temp_result.highest_temp
avg_temp = temp_result.avg_temp

# Print the results
print("Lowest Temperature:", lowest_temp)
print("Highest Temperature:", highest_temp)
print("Average Temperature:", avg_temp)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
tobs_query = session.query(Measurement.tobs).filter(Measurement.station == station_id,
                                                    Measurement.date >= one_year_ago).all()

# Creating df from results
df = pd.DataFrame(tobs_query, columns=['Temperature'])

# Counting occurrences of each temperature
temperature_counts = df['Temperature'].value_counts().sort_index()

# Setting up figure and axes
fig, ax = plt.subplots()

# Creating bar graph
ax.bar(temperature_counts.index, temperature_counts.values)

# Labeling
ax.set_xlabel('Temperature')

ax.set_ylabel('Frequency')

ax.set_title(f'Frequency of Temperature Readings for Station {station_id}')

# Display the graph
plt.show()
```
